## Data Cleaning :


##### Removing unused columns which will not be useful during analysis

Unused Columns -  id, imdb_id, popularity, budget_adj, revenue_adj, homepage, keywords, overview, production_companies, vote_count and vote_average.

```python
# list of columns that needs to be removed
del_col = [ 'id', 'imdb_id', 'popularity', 'budget_adj', 'revenue_adj', 'homepage', 'keywords', 
           'overview', 'production_companies', 'vote_count', 'vote_average']

# deleting the columns from data set
tmdb_data = tmdb_data.drop(del_col, 1)

# checking changes
tmdb_data.head()
```

Checking the structure of data set.

```python
def structure(x):
    rows, cols = x.shape
    print('Total Movies:  {} , Total Columns {} '.format(rows-1, cols)) # total movies = row -1 because first row consists of column details.
```

```python
structure(tmdb_data)
```

##### Removing the duplicate items

Removing the duplicate values in whole data, keeping the first value intact.

```python
tmdb_data.drop_duplicates(keep = 'first' , inplace= True)

structure(tmdb_data)
```

So as per the above result we can clearly see that there was a single duplicate item, since total rows count became one less.

##### Removing rows which consists of '0' value in the revenue or budget column

While going on through the data, few values were 0 in revenue and budget column. So we need to remove/delete those rows as it would 
totally mess up the analysis.

```python
#list of column that needs to be checked
check_row = ['budget', 'revenue']

#replace the value of '0' to NaN of columns given in the list
tmdb_data[check_row] = tmdb_data[check_row].replace(0, np.NaN)

#drop row which conists of NaN values 
tmdb_data.dropna(subset = check_row, inplace = True)

structure(tmdb_data)
```

After analyzing the result we see that only 3853 movies remains. So the other movies had 0 in either revenue or budget column. Thus we removed it.

## Data Analysis:

##### Creating a new column field named profit

```python
#creating new column which will provide the profit of each movie

tmdb_data.insert(2, 'profit($)', tmdb_data['revenue($)'] - tmdb_data['budget($)']) # profit = revenue - budget

tmdb_data.head()
```

```python
# for highest/most and lowest/least values creating a function.

def high_low(col_name):
    
    #higest value (referencing through index value)
    high_val = tmdb_data[col_name].idxmax()
    high_data = pd.DataFrame(tmdb_data.loc[high_val]) # details for high value
    
    #lowest value (referencing through index value)
    low_val = tmdb_data[col_name].idxmin()
    low_data = pd.DataFrame(tmdb_data.loc[low_val]) # details for low value
    
    ans = pd.concat([high_data,low_data], axis = 1)
    
    return ans
```

#### 4. Which genre was most successful ?

```python
# creating a count function for the fields like director, cast and genre
# this function will use '|' as a delimiter to diffrentiate data

def adv_count(column_nam):
    
    data = profit_movies[column_nam].str.cat(sep = '|') #separate the string by '|', each column
    
    data = pd.Series(data.split('|')) # store the data with delimiter '|' 
    
    count = data.value_counts(ascending = False) #highest value first
    
    return count
```

```python
# most successful genre - counts

genres_count = adv_count('genres')

genres_count
```

```python
genres_count.sort_values(ascending = True, inplace = True) # high to low

gen = genres_count.plot.barh() #plotting

gen.set(title = 'Most successful genre') #title

gen.set_xlabel('Density of movies') # x axis

#figure size(width, height)
gen.figure.set_size_inches(12, 9)

#ploting the graph
plt.show()
```

__Most successful genre__ : Comedy [492 successful movies]
